```python
type_counts = data['type'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=type_counts.index, y=type_counts.values,
palette='Set2')
plt.title('Distribution of Content by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Split the 'listed_in' column and count genres
data['genres'] = data['listed_in'].apply(lambda x: x.split(', '))
all_genres = sum(data['genres'], [])
genre_counts = pd.Series(all_genres).value_counts().head(10)
```

```python
data
```

```python
# Plot the most common genres
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.values, y=genre_counts.index,
palette='Set3')
plt.title('Most Common Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
# Count titles by director
top_directors = data['director'].value_counts().head(10)
```

```python
# Plot top directors
plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.values, y=top_directors.index,
palette='Blues_d')
plt.title('Top 10 Directors with the Most Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```

```python
ratings = data['rating'].value_counts().reset_index()
ratings.columns = ['rating', 'count']  # Rename columns
ratings = ratings.sort_values(by='count', ascending=False)  # âœ… Correct line

# Plot the ratings distribution
plt.figure(figsize=(8,5))
plt.bar(ratings['rating'], ratings['count'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel("Rating Types")
plt.ylabel("Rating Frequency")
plt.title('Rating Distribution on Netflix', fontsize=16)
plt.show()
```

```python
# Get the top 10 countries
top_ten_countries = (data['country']
                     .value_counts()
                     .reset_index()
                     .sort_values(by='count', ascending=False)
                     .head(10))  # Select top 10

# Plot the top 10 countries
plt.figure(figsize=(10, 6))
plt.bar(top_ten_countries['country'], top_ten_countries['count'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel("Country")
plt.ylabel("Frequency")
plt.title("Top 10 Countries with Most Content on Netflix", fontsize=14)
plt.show()
```

```python

# Get yearly releases for Movies and TV Shows
yearly_movie_releases = data[data['type'] == 'Movie']['release_year'].value_counts().sort_index()
yearly_series_releases = data[data['type'] == 'TV Show']['release_year'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(yearly_movie_releases.index, yearly_movie_releases.values, label='Movies', marker='o')
plt.plot(yearly_series_releases.index, yearly_series_releases.values, label='TV Shows', marker='s')

plt.xlabel("Years")
plt.ylabel("Frequency of Releases")
plt.title("Yearly Releases of Movies and TV Shows on Netflix", fontsize=14)
plt.legend()
plt.grid(True)
plt.show()

```

```python
popular_movie_genre=data[data['type']=='Movie'].groupby("listed_in").size().sort_values(ascending=False)[:10]
popular_series_genre=data[data['type']=='TVShow'].groupby("listed_in").size().sort_values(ascending=False)[:10]
plt.bar(popular_movie_genre.index, popular_movie_genre.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Genres")
plt.ylabel("Movies Frequency")
plt.suptitle("Top 10 popular genres for movies on Netflix")
plt.show()
```

```python
plt.bar(popular_series_genre.index,popular_series_genre.values)
plt.xticks(rotation=45,ha='right')
plt.xlabel("Genres")
plt.ylabel("TVShowsFrequency")
plt.suptitle("Top10populargenresforTVShowsonNetflix")
plt.show()
```
